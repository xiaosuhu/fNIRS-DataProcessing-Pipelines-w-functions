```python
import os
import shutil
import re

# Define source and destination directories
source_root = "./RawData"
dest_root = "./HH_data_temp"

# Ensure the destination directories exist
# os.makedirs(os.path.join(dest_root, "EN"), exist_ok=True)
# os.makedirs(os.path.join(dest_root, "CH"), exist_ok=True)

# Iterate through first-level subdirectories
for subject_folder in os.listdir(source_root):
    subject_path = os.path.join(source_root, subject_folder)
    
    # Ensure it's a directory
    if not os.path.isdir(subject_path):
        continue
    
    # Extract only the numeric subject ID at the beginning
    match = re.match(r"(\d+)", subject_folder)
    if match:
        subject_id = match.group(1)
    else:
        continue
    
    # Iterate through second-level subdirectories
    for sub_folder in os.listdir(subject_path):
        sub_path = os.path.join(subject_path, sub_folder)
        
        """# Ensure it's a directory
        if not os.path.isdir(sub_path):
            continue
        
        # Determine the category based on the folder name
        if "CHMAPA" in sub_folder:
            category = "CH"
        elif "MAPA" in sub_folder:
            category = "EN"
        else:
            continue"""
        
        # Iterate through .nirs files in the third-level subdirectory
        # for file in os.listdir(sub_path):
        if sub_path.endswith(".nirs"):
                
            # Define subject-specific destination directory using extracted subject ID
            # subject_dest_path = os.path.join(dest_root, category, subject_id)
            subject_dest_path = os.path.join(dest_root, subject_id)
            os.makedirs(subject_dest_path, exist_ok=True)
            
            # Copy the .nirs file
            src_file = sub_path
            dest_file = os.path.join(subject_dest_path, sub_folder)
            shutil.copy2(src_file, dest_file)
            print(f"Copied: {src_file} -> {dest_file}")

print("Extraction complete.")
```
